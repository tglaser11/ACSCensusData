```python
import pandas as pd
```

```python
dp02_data = pd.read_csv('ACS_16_5YR_DP02_with_ann.csv')

dp02_data.drop(['GEO.id','GEO.id2'], 1, inplace=True)
dp02_data.drop([col for col in dp02_data if col.startswith('HC02')], 1, inplace=True)
dp02_data.drop([col for col in dp02_data if col.startswith('HC04')], 1, inplace=True)
dp02_data.to_csv('dp02.csv', index=False)
dp02_data = pd.read_csv('dp02.csv', skiprows=1)
# dp02_data.head()
```

```python
dp03_data = pd.read_csv('ACS_16_5YR_DP03_with_ann.csv')

dp03_data.drop(['GEO.id','GEO.id2'], 1, inplace=True)
dp03_data.drop([col for col in dp03_data if col.startswith('HC02')], 1, inplace=True)
dp03_data.drop([col for col in dp03_data if col.startswith('HC04')], 1, inplace=True)
dp03_data.to_csv('dp03.csv', index=False)
dp03_data = pd.read_csv('dp03.csv', skiprows=1)
# dp03_data.head()
```

```python
dp04_data = pd.read_csv('ACS_16_5YR_DP04_with_ann.csv')

dp04_data.drop(['GEO.id','GEO.id2'], 1, inplace=True)
dp04_data.drop([col for col in dp04_data if col.startswith('HC02')], 1, inplace=True)
dp04_data.drop([col for col in dp04_data if col.startswith('HC04')], 1, inplace=True)
dp04_data.to_csv('dp04.csv', index=False)
dp04_data = pd.read_csv('dp04.csv', skiprows=1)
# dp04_data.head()
```

```python
dp05_data = pd.read_csv('ACS_16_5YR_DP05_with_ann.csv')

dp05_data.drop(['GEO.id','GEO.id2'], 1, inplace=True)
dp05_data.drop([col for col in dp05_data if col.startswith('HC02')], 1, inplace=True)
dp05_data.drop([col for col in dp05_data if col.startswith('HC04')], 1, inplace=True)
dp05_data.to_csv('dp05.csv', index=False)
dp05_data = pd.read_csv('dp05.csv', skiprows=1)
# dp05_data.head()
```

```python
dpall_data = pd.merge(dp02_data, dp03_data, on='Geography', how='outer')
del dp02_data, dp03_data
dpall_data = pd.merge(dpall_data, dp04_data, on='Geography', how='outer')
del dp04_data
dpall_data = pd.merge(dpall_data, dp05_data, on='Geography', how='outer')
del dp05_data
dpall_data.head()
```

```python
dpall_data.to_csv('dpall.csv', index=False)
```

```python
labelfile = open('columns.txt', 'w')
for item in list(dpall_data.columns.values):
    labelfile.write("%s\n" % item)
```
